```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

URL = "https://ko.semrush.com/trending-websites/global/all"
search_url = urllib.request.urlopen(URL).read()
soup = BeautifulSoup(search_url, "html.parser")
rank_df = pd.DataFrame(columns=["도메인", "방문수", "데스크톱 공유(%)", "데스크톱 공유", "MoM", "주요 트래픽 소스"])

sites = soup.select("#open-trends > div > div.Ol5zu8YuQNitkogJRxYc > div.Nu7hm0jz4dvgn1NLDDw9 > section > div.JLbigmpEKM1IcxahHAsl.___SScrollArea_k6o81_gg_ > div > div > table > tbody > tr")

for idx, site in enumerate(sites):
    domain = site.find("a", {"class": "___SBoxInline_1tphm_gg_ NZFMZoXJrDaOPTLvk6JZ"}).get_text()
    visits = site.find("td", {"class": "_valign_top_1bnyc_gg_ _use_primary_1bnyc_gg_ ___SCellRow_1bnyc_gg_ ___SCell_1bnyc_gg_ _align_right_1bnyc_gg_ _valign_top_1bnyc_gg_ _use_primary_1bnyc_gg_ ___SText_6st4r_gg_"}).get_text()
    desktop_percent = site.find("span", {"class": "sKdQ4wd0zfmUujaINE1n"}).get_text()
    desktop = site.find("span", {"class": "z_d4MfXdlvt5KBqKyO54"}).get_text()
    mom = site.find("span", {"class": "OcPtYyYpb2NT9vN1Gy4N V2IsBMTLONaMsboD7A7A"}).get_text()
    traffic_source = site.find("td", {"class": "_valign_top_1bnyc_gg_ _use_primary_1bnyc_gg_ ___SCellRow_1bnyc_gg_ ___SCell_1bnyc_gg_ _align_left_1bnyc_gg_ _valign_top_1bnyc_gg_ _use_primary_1bnyc_gg_ ___SText_6st4r_gg_"}).get_text()
    rank_df.loc[idx + 1] = [domain, visits, desktop_percent, desktop, mom, traffic_source]

rank_df
```
