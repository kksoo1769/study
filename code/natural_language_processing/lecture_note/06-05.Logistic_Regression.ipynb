# Sigmoid Function

$ H(x) = \frac{1}{1 + e^-(wx+b)} = sigmoid(wx+b) = \sigma(wx+b) $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, c="red")
plt.plot((0, 0), (0, 1), linestyle="--")
plt.show()
```

```python
y_1 = sigmoid(0.5 * x)
y_2 = sigmoid(x)
y_3 = sigmoid(2 * x)

plt.plot(x, y_1, c="red")
plt.plot(x, y_2, c="green")
plt.plot(x, y_3, c="blue")
plt.plot((0, 0), (0, 1), linestyle="--")

plt.show() #  w   
```

```python
y_1 = sigmoid(x + 0.5)
y_2 = sigmoid(x + 1)
y_3 = sigmoid(x + 1.5)

plt.plot(x, y_1, c="red")
plt.plot(x, y_2, c="green")
plt.plot(x, y_3, c="blue")
plt.plot((0, 0), (0, 1), linestyle="--")

plt.show() #  b   
```
